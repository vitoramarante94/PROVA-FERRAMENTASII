```python
# Importando Seaborn e matplotlib
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Após selecionei as 4 maiores fornecedoras e plotei um gáfico para cada sobre o total de passagens e os valores dos bilhetes
sns.distplot(df[df['Companhia Aérea'].str.contains('GOL Linhas Aéreas Inteligentes')]['Valor Bilhete'].dropna(), label='GOL')
plt.legend()
```

```python

sns.distplot(df[df['Companhia Aérea'].str.contains('LATAM AIRLINES BRASIL')]['Valor Bilhete'].dropna(), label='LATAM')
plt.legend()
```

```python
sns.distplot(df[df['Companhia Aérea'].str.contains('AZUL LINHAS AÉREAS BRASILEIRA')]['Valor Bilhete'].dropna(), label='AZUL')
plt.legend()
```

```python
sns.distplot(df[df['Companhia Aérea'].str.contains('AVIANCA S.A.')]['Valor Bilhete'].dropna(), label='AVIANCA S.A.')
plt.legend()
```

```python
# Plotei os 4 gráficos sobrepostos para analisar se há alguma diferênça sobre os valores das passagens

sns.distplot(df[df['Companhia Aérea'].str.contains('LATAM AIRLINES BRASIL')]['Valor Bilhete'].dropna(), label='LATAM')
sns.distplot(df[df['Companhia Aérea'].str.contains('AZUL LINHAS AÉREAS BRASILEIRA')]['Valor Bilhete'].dropna(), label='AZUL')
sns.distplot(df[df['Companhia Aérea'].str.contains('GOL Linhas Aéreas Inteligentes')]['Valor Bilhete'].dropna(), label='GOL')
sns.distplot(df[df['Companhia Aérea'].str.contains('AVIANCA S.A.')]['Valor Bilhete'].dropna(), label='AVIANCA S.A.')

plt.legend()
```
